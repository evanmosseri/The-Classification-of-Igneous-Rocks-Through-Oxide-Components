```python
import numpy as np 
import scipy as st 
import matplotlib.pyplot as plt 
import pylab as pl
import csv
from sklearn import neighbors, datasets
import csv
from matplotlib.colors import ListedColormap
import sys
import pandas as pd
import matplotlib.colors as mcolors
import matplotlib.colorbar as colorbar
import matplotlib.axes as maxes
import matplotlib.ticker as mticker
import util

np.set_printoptions(precision=2)

def read_data(filename):
	label_field = 'ROCK NAME'
	labels, feature_mat = None, None
	with open(filename,'rU') as csvfile:
		dialect = csv.Sniffer().sniff(csvfile.read(8024))
		csvfile.seek(0)
		dr = csv.DictReader(csvfile, dialect=dialect)
		data = [row for row in dr]
		labels = np.array([row[label_field] for row in data])
		feature_fields = [field for field in data[0].keys() if field != label_field]
		try:
			feature_mat = np.squeeze(np.array([[float(row[field]) for field in feature_fields] for row in data]))
		except Exception, e:
			print "row[",field,"]=", row[field]
			print "row =", row
			raise e
	return labels, feature_mat, feature_fields

def component(d,n,comp):
    return d[1][n][d[2].index(comp)]
dataURL = "peridotites_clean_complete.csv"
X = read_data(dataURL)
component(X,0,"SIO2")
```

```python
def scatter_matrix(data, rock_labels, rock_names, rock_color, features, title, norm):
    """Customizes pandas scatter_matrix"""
    df = pd.DataFrame(data, columns=features)

    
    fig = plt.figure(figsize=(12,12))
    

    ax = fig.add_subplot(1, 1, 1)

    ax.set_title("my title")
    axes = pd.tools.plotting.scatter_matrix(
        df, alpha=0.2, figsize=(6, 6), ax=ax)

    n = len(features)
    ut = [(r,c) for r in range(n) for c in range(r, n) if r!=c]
    lt = [(r,c) for r in range(n) for c in range(r) if r!=c]
    
    if norm:
        fstr = '%.2f'
    else:
        fstr = '%.5f'
                                                   
    fig.set_size_inches(30,30)
    nl = 10
    ticks = dict()
    for f in features:
        ticks[f] = np.linspace(df[f].min(), df[f].max(), nl)
        

    for r, rf in zip(range(n), features):
        for c, cf in zip(range(n), features):
            sax = axes[r, c]
            
            if (r,c) in ut:
                sax.cla()
                sax.set_axis_off()               
                
            if (r,c) in lt:
                sax.cla()
                obs1 = df[rf].values
                obs2 = df[cf].values
                #rock labels are list of labels of observations
                rock1 = (rock_labels.values==rock_names[0])
                rock2 = (rock_labels.values==rock_names[1])
                rock3 = (rock_labels.values==rock_names[2])
                rock4 = (rock_labels.values==rock_names[3])

                sax.scatter(obs2[rock1], obs1[rock1], s=12, edgecolor='k', color=rock_color[0])
                sax.scatter(obs2[rock2], obs1[rock2], s=12, edgecolor='k', color=rock_color[1])
                sax.scatter(obs2[rock3], obs1[rock3], s=12, edgecolor='k', color=rock_color[2])
                sax.scatter(obs2[rock4], obs1[rock4], s=12, edgecolor='k', color=rock_color[3])
                        
                sax.xaxis.set_major_locator(mticker.FixedLocator(ticks[cf]))
                sax.yaxis.set_major_locator(mticker.FixedLocator(ticks[rf]))
        
            if (r==c):
                sax.cla()
                sax.hist(df[rf], bins=ticks[rf], color="#006699")
            
            if (c==0):
                sax.set_ylabel(features[r])
                sax.yaxis.set_visible(True)
                sax.yaxis.set_major_locator(mticker.LinearLocator(nl))
                sax.yaxis.set_ticklabels(ticks[rf])
                sax.yaxis.set_major_formatter(mticker.FormatStrFormatter(fstr))
                for lab in sax.yaxis.get_ticklabels():
                    lab.set_fontsize(12)                 

                sax.yaxis.get_ticklabels()[0].set_visible(False)
                sax.yaxis.get_ticklabels()[-1].set_visible(False)

                
            if (r==(n-1)):
                sax.set_xlabel(features[c])
                sax.xaxis.set_visible(True)
                sax.xaxis.set_major_locator(mticker.LinearLocator(nl))
                sax.xaxis.set_ticklabels(ticks[cf])
                sax.xaxis.set_major_formatter(mticker.FormatStrFormatter(fstr))
                for lab in sax.xaxis.get_ticklabels():
                    lab.set_rotation(90)
                    lab.set_fontsize(12)
                sax.xaxis.get_ticklabels()[0].set_visible(False)
                sax.xaxis.get_ticklabels()[-1].set_visible(False)
             
            if (r==0) and (c==0):
                sax.set_yticks([])                

    ax.autoscale(tight=True)
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
dt = pd.read_csv(dataURL)
#smatrix = pd.scatter_matrix(dt, diagonal="kde")
varlist = ['SIO2','AL2O3','MGO','CAO']
raw_data = np.vstack(dt[v] for v in varlist)
print raw_data.shape
rock_labels = dt['ROCK NAME']
rock_names = np.unique(rock_labels)
print rock_names
rock_colors = ['r','g','b','orange']
#raw_data = raw_data.reshape(len(raw_data[0]),len(raw_data))
#scam = pd.DataFrame(raw_data.T,columns=varlist)
# pd.scatter_matrix(scam,diagonal="kde")
scatter_matrix(raw_data.T, rock_labels, rock_names, rock_colors, varlist, "title", False)
plt.savefig("figure_2.png")
```
